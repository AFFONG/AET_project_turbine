```python
import math
import numpy as np
import matplotlib.pyplot as plt

work_coeff = 1
flow_coeff = 0.55

r_mean = 0.42  # m

# FOR STAGE 1 TURBINE
r_tip = 0.4865  # m change
r_hub = 0.3535  # m change

blade_height = r_tip - r_hub  # m

# -------------------------
# FUNCTIONS
# -------------------------

def reaction(v_axial, U, alpha_2, alpha_3):  # notes page 45
    return 1 - v_axial / (2 * U) * (
        math.tan(math.radians(alpha_2)) +
        math.tan(math.radians(alpha_3))
    )

def alpha3_to_alpha2(work_coeff, flow_coeff, alpha_3):
    return math.degrees(
        math.atan(work_coeff / flow_coeff + math.tan(math.radians(alpha_3)))
    )

def alpha_to_beta(flow_coeff, alpha):
    return math.degrees(
        math.atan(math.tan(math.radians(alpha)) - 1 / flow_coeff)
    )

# -------------------------
# MEAN RADIUS
# -------------------------

U_mean = 422.75  # m/s
v_axial = flow_coeff * U_mean  # axial velocity

alpha_mean_1 = 0 # change
alpha_mean_3 = 0 # change

alpha_mean_2 = alpha3_to_alpha2(work_coeff, flow_coeff, alpha_mean_3)

beta_mean_2 = alpha_to_beta(flow_coeff, alpha_mean_2)
beta_mean_3 = alpha_to_beta(flow_coeff, alpha_mean_3)

v_tangential_mean_2 = math.tan(math.radians(alpha_mean_2)) * v_axial
v_tangential_mean_3 = math.tan(math.radians(alpha_mean_3)) * v_axial

R_mean = reaction(v_axial, U_mean, alpha_mean_2, alpha_mean_3)

print("----- MEAN RADIUS -----")
print(f"alpha1 = {alpha_mean_1:.2f} deg")
print(f"alpha2 = {alpha_mean_2:.2f} deg")
print(f"alpha3 = {alpha_mean_3:.2f} deg")

print(f"beta2 = {beta_mean_2:.2f} deg")
print(f"beta3 = {beta_mean_3:.2f} deg")

print(f"Tangential velocity at rotor inlet (v_theta2) = {v_tangential_mean_2:.2f} m/s")
print(f"Tangential velocity at rotor outlet (v_theta3) = {v_tangential_mean_3:.2f} m/s")

print(f"Blade speed at mean radius (U_mean) = {U_mean:.2f} m/s")
print(f"Axial velocity (v_axial) = {v_axial:.2f} m/s")

print(f"Degree of Reaction (mean) = {R_mean:.3f} [-]")


# -------------------------
# FUNCTION TO COMPUTE ANY RADIUS
# -------------------------

def radial_properties(r, r_mean, U_mean, v_axial,
                      v_t_mean_2, v_t_mean_3):

    U = r / r_mean * U_mean

    # Free vortex design
    v_t2 = r_mean / r * v_t_mean_2
    v_t3 = r_mean / r * v_t_mean_3


    alpha2 = math.degrees(math.atan(v_t2 / v_axial))
    alpha3 = math.degrees(math.atan(v_t3 / v_axial))

    beta2 = alpha_to_beta(flow_coeff, alpha2)
    beta3 = alpha_to_beta(flow_coeff, alpha3)

    R = reaction(v_axial, U, alpha2, alpha3)

    return U, alpha2, alpha3, beta2, beta3, R

# -------------------------
# TIP CALCULATION
# -------------------------

# Assume axial velocity constant at all radii and stages

U_tip, alpha_tip_2, alpha_tip_3, beta_tip_2, beta_tip_3, R_tip = \
    radial_properties(r_tip, r_mean, U_mean, v_axial,
                      v_tangential_mean_2, v_tangential_mean_3)

print("\n----- TIP -----")
print(f"Blade speed at tip (U_tip) = {U_tip:.2f} m/s")
print(f"alpha_tip_2 = {alpha_tip_2:.2f} deg")
print(f"alpha_tip_3 = {alpha_tip_3:.2f} deg")

print(f"beta_tip_2 = {beta_tip_2:.2f} deg")
print(f"beta_tip_3 = {beta_tip_3:.2f} deg")

print(f"Degree of Reaction (tip) = {R_tip:.3f} [-]")

# -------------------------
# HUB CALCULATION
# -------------------------

U_hub, alpha_hub_2, alpha_hub_3, beta_hub_2, beta_hub_3, R_hub = \
    radial_properties(r_hub, r_mean, U_mean, v_axial,
                      v_tangential_mean_2, v_tangential_mean_3)

print("\n----- HUB -----")
print(f"Blade speed at hub (U_hub) = {U_hub:.2f} m/s")
print(f"alpha_hub_2 = {alpha_hub_2:.2f} deg")
print(f"alpha_hub_3 = {alpha_hub_3:.2f} deg")

print(f"beta_hub_2 = {beta_hub_2:.2f} deg")
print(f"beta_hub_3 = {beta_hub_3:.2f} deg")

print(f"Degree of Reaction (hub) = {R_hub:.3f} [-]")
```

```python
# =====================================================
# ZWEIFEL COEFFICIENT SPECIFICATION
# =====================================================

# specify Zweifel numbers
zw_mean = 0.8
zw_tip = 0.8
zw_hub = 0.8


# =====================================================
# FUNCTION: ZW -> PITCH TO CHORD RATIO
# =====================================================

def zw_to_sc(zw, angle_in, angle_out):

    angle_in = math.radians(angle_in)
    angle_out = math.radians(angle_out)

    return zw * 0.5 / abs(
        (math.cos(angle_out)**2) *
        abs(math.tan(angle_out) - math.tan(angle_in))
    )


# =====================================================
# MEAN RADIUS
# =====================================================

# stator uses absolute angles
sc_stator_mean = zw_to_sc(zw_mean, alpha_mean_1, alpha_mean_2)

# rotor uses relative angles
sc_rotor_mean = zw_to_sc(zw_mean, beta_mean_2, beta_mean_3)


# =====================================================
# TIP
# =====================================================

# stator (absolute angles)
sc_stator_tip = zw_to_sc(zw_tip, alpha_mean_1, alpha_tip_2)

# rotor (relative angles)
sc_rotor_tip = zw_to_sc(zw_tip, beta_tip_2, beta_tip_3)


# =====================================================
# HUB
# =====================================================

# stator
sc_stator_hub = zw_to_sc(zw_hub, alpha_mean_1, alpha_hub_2)

# rotor
sc_rotor_hub = zw_to_sc(zw_hub, beta_hub_2, beta_hub_3)


# =====================================================
# PRINT RESULTS
# =====================================================

print("\n========== ZWEIFEL COEFFICIENT ==========")

print(f"Zw (mean) = {zw_mean}")
print(f"Zw (tip)  = {zw_tip}")
print(f"Zw (hub)  = {zw_hub}")

print("\n----- PITCH TO CHORD RATIO (s/c) -----")

print(f"Stator (mean) s/c = {sc_stator_mean:.3f} [-]")
print(f"Rotor  (mean) s/c = {sc_rotor_mean:.3f} [-]")

print(f"Stator (tip)  s/c = {sc_stator_tip:.3f} [-]")
print(f"Rotor  (tip)  s/c = {sc_rotor_tip:.3f} [-]")

print(f"Stator (hub)  s/c = {sc_stator_hub:.3f} [-]")
print(f"Rotor  (hub)  s/c = {sc_rotor_hub:.3f} [-]")
```

```python
# OUTPUT RANGES fo number of blades, chord and pitch

# =====================================================
# BLADE GEOMETRY FROM ASPECT RATIO
# =====================================================

blade_height = r_tip - r_hub

# aspect ratio limits
AR_STATOR_MIN = 2
AR_STATOR_MAX = 4

AR_ROTOR_MIN = 2
AR_ROTOR_MAX = 4


def chord_range(height, AR_min, AR_max):
    """Return min and max chord based on aspect ratio."""
    c_min = height / AR_max
    c_max = height / AR_min
    return c_min, c_max


def blade_geometry_range(radius, sc_ratio, chord_min, chord_max):
    """Compute pitch and blade count ranges."""

    pitch_min = sc_ratio * chord_min
    pitch_max = sc_ratio * chord_max

    N_max = round(2 * math.pi * radius / pitch_min)
    N_min = round(2 * math.pi * radius / pitch_max)

    return pitch_min, pitch_max, N_min, N_max


def intersect_blade_range(N_mean_min, N_mean_max,
                          N_tip_min, N_tip_max,
                          N_hub_min, N_hub_max):
    """Find blade count range valid for mean, tip and hub."""
    
    N_lower = max(N_mean_min, N_tip_min, N_hub_min)
    N_upper = min(N_mean_max, N_tip_max, N_hub_max)

    return N_lower, N_upper


# =====================================================
# STATOR GEOMETRY
# =====================================================

c_stator_min, c_stator_max = chord_range(
    blade_height,
    AR_STATOR_MIN,
    AR_STATOR_MAX
)

pitch_stator_min, pitch_stator_max, \
N_stator_min, N_stator_max = blade_geometry_range(
    r_mean,
    sc_stator_mean,
    c_stator_min,
    c_stator_max
)

# tip
pitch_stator_tip_min, pitch_stator_tip_max, \
N_stator_tip_min, N_stator_tip_max = blade_geometry_range(
    r_tip,
    sc_stator_tip,
    c_stator_min,
    c_stator_max
)

# hub
pitch_stator_hub_min, pitch_stator_hub_max, \
N_stator_hub_min, N_stator_hub_max = blade_geometry_range(
    r_hub,
    sc_stator_hub,
    c_stator_min,
    c_stator_max
)

# final stator blade range
N_stator_lower, N_stator_upper = intersect_blade_range(
    N_stator_min, N_stator_max,
    N_stator_tip_min, N_stator_tip_max,
    N_stator_hub_min, N_stator_hub_max
)


# =====================================================
# ROTOR GEOMETRY
# =====================================================

c_rotor_min, c_rotor_max = chord_range(
    blade_height,
    AR_ROTOR_MIN,
    AR_ROTOR_MAX
)

pitch_rotor_min, pitch_rotor_max, \
N_rotor_min, N_rotor_max = blade_geometry_range(
    r_mean,
    sc_rotor_mean,
    c_rotor_min,
    c_rotor_max
)

# tip
pitch_rotor_tip_min, pitch_rotor_tip_max, \
N_rotor_tip_min, N_rotor_tip_max = blade_geometry_range(
    r_tip,
    sc_rotor_tip,
    c_rotor_min,
    c_rotor_max
)

# hub
pitch_rotor_hub_min, pitch_rotor_hub_max, \
N_rotor_hub_min, N_rotor_hub_max = blade_geometry_range(
    r_hub,
    sc_rotor_hub,
    c_rotor_min,
    c_rotor_max
)

# final rotor blade range
N_rotor_lower, N_rotor_upper = intersect_blade_range(
    N_rotor_min, N_rotor_max,
    N_rotor_tip_min, N_rotor_tip_max,
    N_rotor_hub_min, N_rotor_hub_max
)


# =====================================================
# PRINT RESULTS
# =====================================================

print("\n========== BLADE GEOMETRY ==========")

print(f"Blade height = {blade_height:.4f} m")

print("\n--- STATOR ---")
print(f"Aspect ratio range (AR) = {AR_STATOR_MIN} – {AR_STATOR_MAX}")
print(f"s/c ratio at mean radius = {sc_stator_mean:.3f}")
print(f"Chord range = {c_stator_min:.4f} – {c_stator_max:.4f} m")
print(f"Pitch range = {pitch_stator_min:.4f} – {pitch_stator_max:.4f} m")
print(f"Mean radius blades ≈ {N_stator_min} – {N_stator_max}")
print(f"Tip radius blades  ≈ {N_stator_tip_min} – {N_stator_tip_max}")
print(f"Hub radius blades  ≈ {N_stator_hub_min} – {N_stator_hub_max}")
print(f"VALID stator blade range = {N_stator_lower} – {N_stator_upper}")

print("\n--- ROTOR ---")
print(f"Aspect ratio range (AR) = {AR_ROTOR_MIN} – {AR_ROTOR_MAX}")
print(f"s/c ratio at mean radius = {sc_rotor_mean:.3f}")
print(f"Chord range = {c_rotor_min:.4f} – {c_rotor_max:.4f} m")
print(f"Pitch range = {pitch_rotor_min:.4f} – {pitch_rotor_max:.4f} m")
print(f"Mean radius blades ≈ {N_rotor_min} – {N_rotor_max}")
print(f"Tip radius blades  ≈ {N_rotor_tip_min} – {N_rotor_tip_max}")
print(f"Hub radius blades  ≈ {N_rotor_hub_min} – {N_rotor_hub_max}")
print(f"VALID rotor blade range = {N_rotor_lower} – {N_rotor_upper}")
```

```python
def blade_geometry(N_stator, N_rotor, sc_stator, sc_rotor,
                   r_hub, r_mean, r_tip, blade_height):
    """
    Compute pitch, chord, and aspect ratio at hub, mean, tip for stator and rotor.
    
    sc_stator, sc_rotor: dict with keys 'hub','mean','tip'
    """
    results = {'stator': {}, 'rotor': {}}
    
    for row, N, sc in [('stator', N_stator, sc_stator), ('rotor', N_rotor, sc_rotor)]:
        results[row]['hub']  = {}
        results[row]['mean'] = {}
        results[row]['tip']  = {}
        
        for pos, r in [('hub', r_hub), ('mean', r_mean), ('tip', r_tip)]:
            pitch = 2 * math.pi * r / N
            chord = pitch / sc[pos]
            AR = blade_height / chord
            results[row][pos] = {'pitch': pitch, 'chord': chord, 'AR': AR}
    
    return results

# DEFINE NUMBER OF STATOR BLADE AND ROTOR BLADES
N_stator = 59
N_rotor = 53

sc_stator = {'hub': sc_stator_hub, 'mean': sc_stator_mean, 'tip': sc_stator_tip}
sc_rotor  = {'hub': sc_rotor_hub,  'mean': sc_rotor_mean,  'tip': sc_rotor_tip}

geometry = blade_geometry(N_stator, N_rotor,
                          sc_stator=sc_stator, sc_rotor=sc_rotor,
                          r_hub=r_hub, r_mean=r_mean, r_tip=r_tip,
                          blade_height=blade_height)

# Print nicely
print(f"Number of Stator Blades: {N_stator}")
print(f"Number of Rotor Blades: {N_rotor}")
print(f"stator-rotor blade ratio: {N_stator/N_rotor:.2f}")

for row in ['stator','rotor']:
    print(f"\n--- {row.upper()} ---")
    for pos in ['hub','mean','tip']:
        g = geometry[row][pos]
        print(f"{pos.capitalize()}: pitch={g['pitch']:.4f} m, chord={g['chord']:.4f} m, AR={g['AR']:.2f}")
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt

# ==========================================
# 1. INPUT PARAMETERS
# ==========================================
chord_length = 0.0467  # Meters
t_max_target = 0.15    # 15% thickness

# The code is basically mapping the NACA profile (zero camber) to a parobolic camber that is defined with inlet and outlet metal angles

# Angles for Mean station
chi_in = -21.31         
chi_out = -68.06      
gamma = -44.68        

# NACA 63012A Data (Normalised 0-1)
naca_data = np.array([
    [0.0000, 0.00000], [0.0050, 0.00973], [0.0075, 0.01173], [0.0125, 0.01492],
    [0.0250, 0.02078], [0.0500, 0.02895], [0.0750, 0.03504], [0.1000, 0.03994],
    [0.1500, 0.04747], [0.2000, 0.05287], [0.2500, 0.05664], [0.3000, 0.05901],
    [0.3500, 0.05995], [0.4000, 0.05957], [0.4500, 0.05792], [0.5000, 0.05517],
    [0.5500, 0.05148], [0.6000, 0.04700], [0.6500, 0.04186], [0.7000, 0.03621],
    [0.7500, 0.03026], [0.8000, 0.02426], [0.8500, 0.01826], [0.9000, 0.01225],
    [0.9500, 0.00625], [1.0000, 0.00025]
])

# ==========================================
# 2. GEOMETRY GENERATION LOGIC
# ==========================================

def rotate_coords(x, y, angle_deg):
    """Rotates coordinates around the origin (0,0)."""
    rad = math.radians(angle_deg)
    xr = x * math.cos(rad) - y * math.sin(rad)
    yr = x * math.sin(rad) + y * math.cos(rad)
    return xr, yr

def generate_blade_geometry(chord, naca_coords, c_in, c_out, g, thickness_ratio):
    x_naca = naca_coords[:, 0]
    yt = naca_coords[:, 1] * (thickness_ratio / 0.12)
    
    chi_in_rel = math.radians(c_in - g)
    B = math.tan(chi_in_rel)
    A = -B 
    
    xu, yu, xl, yl, xc_line, yc_line = [], [], [], [], [], []
    
    for i in range(len(x_naca)):
        x = x_naca[i]
        yc = (A * x**2 + B * x) * chord
        xc = x * chord
        theta = math.atan(2 * A * x + B)
        
        # Local offsets
        xu.append(xc - (yt[i] * chord * math.sin(theta)))
        yu.append(yc + (yt[i] * chord * math.cos(theta)))
        xl.append(xc + (yt[i] * chord * math.sin(theta)))
        yl.append(yc - (yt[i] * chord * math.cos(theta)))
        xc_line.append(xc)
        yc_line.append(yc)
        
    # Rotate all to Global Axial Frame
    xu_r, yu_r = rotate_coords(np.array(xu), np.array(yu), g)
    xl_r, yl_r = rotate_coords(np.array(xl), np.array(yl), g)
    xc_r, yc_r = rotate_coords(np.array(xc_line), np.array(yc_line), g)
        
    return xu_r, yu_r, xl_r, yl_r, xc_r, yc_r

# ==========================================
# 3. OUTPUT & PLOTTING (Axial Orientation)
# ==========================================

xu, yu, xl, yl, xc, yc = generate_blade_geometry(chord_length, naca_data, chi_in, chi_out, gamma, t_max_target)

plt.figure(figsize=(12, 8))

# 1. Profile and Camber
plt.plot(xu, yu, 'r-', linewidth=2, label='Suction Side')
plt.plot(xl, yl, 'b-', linewidth=2, label='Pressure Side')
plt.plot(xc, yc, 'k--', linewidth=1.2, label='Camber Line')

# 2. Chord Line (Staggered relative to Axial)
c_line_x, c_line_y = rotate_coords(np.array([0, chord_length]), np.array([0, 0]), gamma)
plt.plot(c_line_x, c_line_y, 'g:', linewidth=2, label=f'Chord Line (Stagger={gamma:.1f}°)')

# 3. Axial Reference Line
plt.axhline(0, color='black', linewidth=0.8, alpha=0.3, label='Axial Direction')

# 4. Metal Angle Indicators
vec_len = chord_length * 0.15

# Inlet Vector
plt.arrow(xc[0], yc[0], math.cos(math.radians(chi_in))*vec_len, math.sin(math.radians(chi_in))*vec_len, 
          color='blue', head_width=0.001, label=f'Inlet Metal ({chi_in}°)')
# Outlet Vector (pointing away from TE)
plt.arrow(xc[-1], yc[-1], -math.cos(math.radians(chi_out))*vec_len, -math.sin(math.radians(chi_out))*vec_len, 
          color='purple', head_width=0.001, label=f'Outlet Metal ({chi_out}°)')

# Formatting
plt.axis('equal')
plt.grid(True, linestyle=':', alpha=0.5)
plt.title(f"T6 Turbine Blade: Axial Frame Orientation\nAxial Chord (Cx) = {chord_length * math.cos(math.radians(gamma)) * 1000:.2f} mm")
plt.xlabel("Axial Direction (X)")
plt.ylabel("Tangential Direction (Y)")
plt.legend(loc='best', fontsize='small')

plt.show()

# OUTPUT CSV
np.text(np.column_stack(([xu, yu], [xl, yl]))


```
